```python
# tenary operator: egy soros if-else
# szintaxis: "igaz ág" if "feltételvizsgálat" else "hamis ág"
n = 6
res = "Páros" if n % 2 == 0 else "Páratlan"
print(res)
```

```python
# szintaxis: "igaz ág" if "feltétel" else "hamis ág" if "feltétel" else "hamis ág"
n = -6
result = "Pozitív" if n > 0 else "Negatív" if n < 0 else "Nulla"
print(result)
```

```python
# más függvényeken belül is működik, például a print()-en belül
a = 19
b = 20
print("a nagyobb mint b" if a > b else "b nem kisebb a-nál")
```
