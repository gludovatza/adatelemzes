```python
import numpy as np

adatok = np.random.uniform(5, 25, size=56)   # 56 mérés
print(adatok.shape)  # (56,)

# Feladat: alakítsd 7 nap × 8 mérés formába
napokra = adatok.reshape(7, 8)

# Számítsd ki minden nap átlagát:
napi_atlag = napokra.mean(axis=1)
print(napi_atlag)
```

```python
pixels = np.arange(36)  # 1D reprezentáció
print(pixels.shape)     # (36,)

# Feladat: alakítsd vissza 6×6-os képpé
img = pixels.reshape(6, 6)
print(img)
```

```python
dobasok = np.random.randint(1, 7, size=100)

# Feladat: 10×10-as mátrix
tabla = dobasok.reshape(10, 10)

# Átlag minden sorban (kísérletenként)
atlagok = tabla.mean(axis=1)
print(atlagok)
```

```python
adatok = np.random.randint(0, 100, size=15)

# Feladat: 3 szenzor × 5 mérés
matrix = adatok.reshape(3, 5)
print(matrix)

# Szenzoronkénti maximum:
maxok = matrix.max(axis=1)
print(maxok)
```

```python
arr = np.array([0, 1, 2, 0, 3])
print(np.count_nonzero(arr))
mask = arr > 1
print(np.count_nonzero(mask))   # hány elem nagyobb mint 1
# Alternatíva: sum() logikai maszkra --> így a "True" elemeket számolja meg (összegzi)
print(np.sum(mask))
```

```python
fogy = np.random.uniform(0.2, 1.0, size=720)  # kWh/óra
print(fogy.shape)  # (720,)

# Feladat: 30×24-es mátrix (nap × óra)
napok = fogy.reshape(30, 24)

# Napi összes fogyasztás
napi_osszeg = napok.sum(axis=1)
print(napi_osszeg)
```

```python
import matplotlib.pyplot as plt

# numpy-os kód
np.random.seed(0)
adatok = np.random.uniform(5, 25, size=56)   # 56 mérés (7 nap * 8 mérés/nap)
napokra = adatok.reshape(7, 8)
napi_atlag = napokra.mean(axis=1)

# megjelenítés matplotlib-bel
plt.figure(figsize=(8, 5))
plt.bar(range(1, 8), napi_atlag)
plt.title("Heti hőmérséklet – napi átlagok")
plt.xlabel("Nap sorszáma (1..7)")
plt.ylabel("Átlaghőmérséklet")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

pixels = np.arange(36)  # 0..35
img = pixels.reshape(6, 6)

plt.figure(figsize=(5, 5))
plt.imshow(img)
plt.title("6×6 kép (imshow)")
plt.xlabel("Oszlop")
plt.ylabel("Sor")
plt.colorbar(label="Intenzitás")
plt.tight_layout()
plt.show()
```

```python
rng = np.random.default_rng(42)
dobasok = rng.integers(1, 7, size=100)  # 1..6
tabla = dobasok.reshape(10, 10)
sor_atlagok = tabla.mean(axis=1)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), sor_atlagok, marker="o")
plt.title("Kísérletek (sorok) átlagai – 10×10 kockadobás")
plt.xlabel("Kísérlet sorszáma (1..10)")
plt.xticks(np.arange(1, 11))
plt.ylabel("Átlag dobott érték")
plt.tight_layout()
plt.show()
```

```python
np.random.seed(1)
adatok = np.random.randint(0, 100, size=15)
matrix = adatok.reshape(3, 5)  # 3 szenzor, 5 időpont

plt.figure(figsize=(6, 4))
plt.imshow(matrix, aspect="auto")
plt.title("Szenzoradatok (3×5) – hőtérkép")
plt.xlabel("Időpont index (0..4)")
plt.ylabel("Szenzor index (0..2)")
plt.yticks([0, 1, 2])
plt.colorbar(label="Mért érték")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
rng = np.random.default_rng(7)
fogy = rng.uniform(0.2, 1.0, size=720)  # 720 óra = 30 nap * 24 óra
napok = fogy.reshape(30, 24)
napi_osszeg = napok.sum(axis=1)

plt.figure(figsize=(9, 5))
plt.plot(range(1, 31), napi_osszeg, marker="o")
plt.title("Havi energiafogyasztás – napi összeg (kWh)")
plt.xlabel("Nap sorszáma (1..30)")
plt.ylabel("Összes fogyasztás (kWh)")
plt.tight_layout()
plt.show()
```
