```python
# Syntaxis
# new_list = [expression for item in iterable if condition]

# expression: is the operation you want to perform on each element.
# item: is a variable that will be assigned to each element in the iterable.
# iterable: is a sequence of items.
# condition: (optional) is a boolean expression that is evaluated for each item
```

```python
# szöveges lista definiálása 
languages = ['python', 'java', 'php', 'go']
```

```python
# szóhosszúságok kigyűjtése for ciklussal
lengths = []
for l in languages:
    lengths.append(len(l))
print(lengths)
```

```python
# szóhosszúságok kigyűjtése list comprehension megoldással
lengths = [len(x) for x in languages]
print(lengths)
```

```python
# p kezdőbetűs elemek kiválasztása for ciklussal
filtered_list = []
for x in languages:
    if x.startswith('p'):
        filtered_list.append(x)
print(filtered_list)
```

```python
# p kezdőbetűs elemek kiválasztása list comprehension megoldással
filtered_list = [x for x in languages if x.startswith('p')]
print(filtered_list)
```

```python
# lista létrehozása "comprehension"-nel
even_numbers = [x for x in range(1,11) if x % 2 == 0]
print(even_numbers)
```

```python
# ciklus egy sorban
my_list = [2, 3, 4, 5]
new_list = [ite * 2 for item in my_list]
print(new_list)
```

```python
# ciklus és feltételes elágazás együtt
my_list = [5, 11, 20, 18, 25]
new_list = [item for item in my_list if item % 5 == 0]
print(new_list)
```

```python
# ciklus és feltételes elágazás együtt (megfordítva)
new_list = [True if item % 2 == 0 else False for item in my_list]
print(new_list)
```

```python
new_list = []
list1 = [5, 8]
list2 = [1, 2, 3]
for x in list1:
    for y in list2:
        new_list.append(x * y)
print(new_list)
```

```python
# egymásba ágyazott ciklusok
list1 = [5, 8]
list2 = [1, 2, 3]
new_list = [x * y for x in list1 for y in list2 ]
print(new_list)
```

```python
# egymásba ágyazott ciklusok + feltételes elágazás
new_list = [x * y for x in list1 for y in list2 if x % 2 == 0]
print(new_list)
```

```python
# egymásba ágyazott ciklusok + feltételes elágazás több feltétellel
new_list = [x * y for x in list1 for y in list2 if x % 2 == 0 and y % 2 != 0]
print(new_list)
```
