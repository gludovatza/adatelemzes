```python
# 1. feladat — Változó árak rendezése
prices = np.array([3200, 1500, 4500, 2990, 1500, 5100])

# a) Rendezés np.sort()
sorted_prices = np.sort(prices)
print(sorted_prices)

# b) In-place rendezés
prices_copy = prices.copy()
prices_copy.sort()
print(prices_copy)

# c) Legolcsóbb és legdrágább index
idx_sorted = np.argsort(prices) # argsort: rendezi az elemeket és megmondja, hogy hol lesz az adott elem indexe a rendezett tömbben
print("Ennek a tömbnek a legelső eleme a legolcsóbb termék, az utolsó eleme a legdrágább termék:", idx_sorted)
print("Ez viszont nem veszi figyelembe, hogy ha azonos a termékeknek az ára, akkor több legolcsóbb és/vagy legdrágább termék is van.")
cheapest_idx = idx_sorted[0]
most_expensive_idx = idx_sorted[-1]
print(cheapest_idx, most_expensive_idx)
```

```python
# 3. feladat — Adatok rendezése név és életkor szerint (structured array)
data = np.array([
    ("Anna", 22),
    ("Péter", 25),
    ("János", 20),
    ("Anna", 21)
], dtype=[("name", "U10"), ("age", "i4")])

# a) Életkor szerint rendezés
by_age = np.sort(data, order="age")
print(by_age)
# VAGY (egy kicsit bonyolultabban):
idx = np.argsort(data['age'])
data_sorted = data[idx]
print(data_sorted)

# b) Először név, azon belül életkor
by_name_age = np.sort(data, order=["name", "age"])
print(by_name_age)

# c) Legidősebb személy
oldest = data[np.argmax(data["age"])]
print("Legidősebb:", oldest)
youngest = data[np.argmin(data["age"])]
print("Legfiatalabb:", youngest)
```

```python
# 4. feladat — Fizetések rendezése több kritérium alapján
employees = np.array([
    ("Anna", "Nagy", 4000),
    ("Gábor", "Kovács", 3800),
    ("Péter", "Németh", 5000),
    ("Anna", "Kis", 4200),
], dtype=[("firstname","U10"),("lastname","U10"),("salary","i4")]) # vezetéknevesen lenne a legjobb!

# a) Név szerint
by_name = np.sort(employees, order="lastname")
print(by_name)

# b) Név, majd fizetés szerint
by_name_salary = np.sort(employees, order=["lastname", "salary"])
print(by_name_salary)

# c) Legmagasabb fizetésű Anna
annas = employees[employees["firstname"] == "Anna"]
print("Ők az Annák:", annas)
best_anna = annas[np.argmax(annas["salary"])]
print("Ő a legjobban kereső Anna:", best_anna)
```

```python
# Rendezés fizetés (salary) szerint
sorted_employees = np.sort(employees, order="salary")

# Külön kivesszük a mezőket a rendezett tömbből
first = sorted_employees["firstname"]
last = sorted_employees["lastname"]
names_sorted = np.char.add(first, " ") # keresztnév után szóköz
names_sorted = np.char.add(names_sorted, last) # keresztnév + szóköz + vezetéknév
print(names_sorted)

salary_sorted = sorted_employees["salary"]

# Bar chart
plt.figure()
plt.bar(range(len(names_sorted)), salary_sorted)
plt.xticks(range(len(names_sorted)), names_sorted, rotation=45)
plt.title("Fizetések növekvő sorrendben (structured array)")
plt.xlabel("Név")
plt.ylabel("Fizetés")
plt.show()
```

```python
# 5. feladat — Tömb véletlen összekeverése
arr = np.array([1, 2, 3, 4, 5, 6])

# a) Shuffle --> a véletlen mindig másképp szórja szét a számokat
shuffled = arr.copy()
np.random.shuffle(shuffled)
print("a)", shuffled)

# b) Rendezett változat
sorted_shuffled = np.sort(shuffled)
print("b)", sorted_shuffled)

# c) Rendező indexek
idx = np.argsort(shuffled)
print("c)", idx)
```

```python
# 6. feladat — Legkisebb 3 érték kiválasztása
x = np.array([12, 4, 22, 9, 1, 15, 7])
print("A tömb kezdetben:", x)

# a) Három legkisebb érték
smallest3_vals = np.sort(x)[:3]
print("Három legkisebb érték:", smallest3_vals)

# b) Ezek indexei
smallest3_idx = np.argsort(x)[:3]
print("Három legkisebb érték indexe:", smallest3_idx)

# c) Három legkisebb érték törlése
# maszkoló tömb, amely ugyanolyan méretű mint az x és mindenhol True kezdetben
mask = np.ones(len(x), dtype=bool) 
print("mask kezdetben:", mask)
mask[smallest3_idx] = False
print("mask módosítva:", mask)
x_without_smallest = x[mask]
print("Három legkisebb elem törlése után a tömb:", x_without_smallest)
```

```python
plt.figure()
plt.plot(x, marker='o')
plt.scatter(smallest3_idx, smallest3_vals, marker='x')   # legkisebb értékek
plt.title("Legkisebb 3 érték kiemelve")
plt.xlabel("Index")
plt.ylabel("Érték")
plt.show()
```

```python
# 7. feladat — Idősor rendezése érték szerint (argsort)
energy = np.array([120, 135, 90, 200, 160, 140])
time = np.array(["08:00","09:00","10:00","11:00","12:00","13:00"])

# a) Energia rendezése
sorted_energy = np.sort(energy)
print(sorted_energy)

# b) Időpontok energia szerint rendezve
idx = np.argsort(energy)
times_sorted = time[idx]
print(times_sorted)

# c) 2 legnagyobb energia időpontja
top2_idx = np.argsort(energy)[-2:]
top2_times = time[top2_idx]
print(top2_times)
```

```python
plt.figure()
plt.plot(time, energy, marker='o')
plt.title("Energiafogyasztás idő szerint")
plt.xlabel("Idő")
plt.ylabel("kWh")
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. feladat — Többdimenziós rangsorolás
b = np.array([
    [10, 40, 20],
    [30, 5, 15]
])

# a) Sori rendező indexek
row_sort_idx = np.argsort(b, axis=1)
print(row_sort_idx)

# b) Oszlop csökkenő rendező indexek
col_sort_idx = np.argsort(-b, axis=0)
print(col_sort_idx)

# c) Mátrix 1D-re lapítva + ranking
ranking = np.argsort(np.argsort(b.flat))
print(ranking)
```

```python
plt.figure()
plt.scatter(b.flat, ranking)
plt.title("Mátrix elemeinek rangsora")
plt.xlabel("Érték")
plt.ylabel("Rang")
plt.show()
```

```python
# 9. feladat — Tanulók pontszámai (ranking)
scores = np.array([85, 92, 70, 88, 92, 75])

# a) Rangok
rank = np.argsort(np.argsort(scores))
print(rank)

# b) Top 3 érték indexei
top3_idx = np.argsort(scores)[-3:]   # UTOLSÓ 3 a rendezett indexek közül
print(top3_idx)

# c) Holtversenyben első tanulók
max_score = scores.max()
best_indices = np.where(scores == max_score)[0]
print(best_indices)
```

```python
plt.figure()
plt.scatter(scores, rank)
plt.title("Pontszámok rangsora")
plt.xlabel("Pontszám")
plt.ylabel("Rang")
plt.show()
```

```python
# 10. feladat — Hiányzó értékek kezelése rendezés előtt
values = np.array([30, np.nan, 10, 50, np.nan, 20])

# a) NaN --> átlag
mean_val = np.nanmean(values)
clean1 = np.where(np.isnan(values), mean_val, values)
sorted1 = np.sort(clean1)
print(sorted1)

# b) NaN --> nagy szám
clean2 = np.where(np.isnan(values), 1e9, values)
sorted2 = np.sort(clean2)
print(sorted2)

# c) Igazi rendezett sorrend (NaN-ok hátul)
nan_mask = np.isnan(values)
real_sorted = np.hstack([
    np.sort(values[~nan_mask]),
    values[nan_mask]
])
print(real_sorted)
```

```python
plt.figure()
plt.plot(clean1, marker='o')
plt.plot(sorted1, marker='o')
plt.title("NaN-ok helyettesítése és rendezése")
plt.xlabel("Index")
plt.ylabel("Érték")
plt.show()
```
