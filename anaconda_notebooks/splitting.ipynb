```python
# 1. feladat Heti adatok napokra bontása

import numpy as np

adatok = np.random.uniform(10, 30, size=336)  # 7 nap * 48 félóra, Véletlenszámokat generál egyenletes eloszlásból (uniform distribution): Minden érték ugyanolyan eséllyel fordul elő a megadott tartományon belül. [alsó határ, felső határ)
napok = np.array_split(adatok, 7)             # 7 részre bontás

napi_atlagok = [nap.mean() for nap in napok]
legmelegebb = np.argmax(napi_atlagok) # legnagyobb érték hol található? indexet ad vissza

print("Napi átlagok:", napi_atlagok)
print("Legmagasabb átlagú nap indexe:", legmelegebb)
```

```python
# 1. feladat kiegészítése diagramos ábrázolással:
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(range(1, 8), napi_atlagok)
plt.title("Napi átlaghőmérsékletek")
plt.xlabel("Nap")
plt.ylabel("Átlag (°C)")
plt.tight_layout()
plt.show()
```

```python
# 2. feladat: termékek 3 polcra bontása

prices = np.random.uniform(1000, 5000, size=30)   # 30 ár
polcok = np.array_split(prices, 3)                # 3 részre

polc_atlagok = [p.mean() for p in polcok]
legdragabb_polc = np.argmax(polc_atlagok)

print("Polcátlagok:", polc_atlagok)
print("Legdrágább polc indexe:", legdragabb_polc)
```

```python
# 2. feladat kiegészítve diagramos ábrázolással

plt.figure(figsize=(7,5))
plt.bar(["Polc 1", "Polc 2", "Polc 3"], polc_atlagok)
plt.title("Polconkénti átlagár")
plt.ylabel("Ár (Ft)")
plt.tight_layout()
plt.show()
```

```python
# 3. Feladat - 4000 adat → 4 CPU-ra szétosztás + vizualizáció

adatok = np.random.rand(4000)
cpu = np.array_split(adatok, 4)   # 4 rész

feldolg_idok = [c.sum() for c in cpu]
legtobb_munka = np.argmax(feldolg_idok)

print("CPU terhelések:", feldolg_idok)
print("Legjobban terhelt mag indexe:", legtobb_munka)
```

```python
# 3. feladat kiegészítve diagramos megjelenítéssel:

plt.figure(figsize=(8,4))
plt.bar(["CPU1", "CPU2", "CPU3", "CPU4"], feldolg_idok)
plt.title("CPU-k terhelése")
plt.ylabel("Összeg (szimulált munka)")
plt.tight_layout()
plt.show()
```

```python
# 4. Feladat – Szenzor-mátrix soronkénti bontása

matrix = np.random.randint(0, 100, size=(10, 5))

sorok = np.vsplit(matrix, 10)     # 10 db 1×5 sor

sorosszegek = [s.sum() for s in sorok]
legnagyobb = np.argmax(sorosszegek)

print("Sorösszegek:", sorosszegek)
print("Legnagyobb összegű sor indexe:", legnagyobb)
```

```python
# 4. feladat kiegészítve diagramos ábrázolással

plt.figure(figsize=(9,4))
plt.plot(sorosszegek, marker="o")
plt.title("Soronkénti mért összérték (10 időpont)")
plt.xlabel("Időpont index")
plt.ylabel("Szenzorösszeg")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()
```

```python
# 5. Feladat – 2D táblázat oszlopokra bontása

import matplotlib.pyplot as plt

data = np.random.rand(100, 4)   # 100 sor, 4 oszlop
cols = np.hsplit(data, 4)       # 4 külön oszlop

for i, col in enumerate(cols):
    plt.figure(figsize=(8,3))
    plt.plot(col, label=f"{i}. oszlop")
    plt.title(f"{i}. oszlop értékeinek alakulása")
    plt.xlabel("Sorindex")
    plt.ylabel("Érték")
    plt.grid(True, linestyle=":")
    plt.tight_layout()
    plt.show()

# Ingadozás = max - min
ingadozasok = [c.max() - c.min() for c in cols]
print("Ingadozások:", ingadozasok)
print("Legnagyobb ingadozás oszlopindexe:", np.argmax(ingadozasok))
```

```python
# 6. Feladat – Hang-fájl 1 mp-es szeletekre bontva - kiegészítve diagramos ábrázolással

signal = np.random.uniform(-1, 1, size=8000)  # hang hullámforma

szeletek = np.array_split(signal, 8)          # 8 darab 1 mp-es rész

# Teljesítmény = négyzetösszeg
teljesitmenyek = [np.sum(s**2) for s in szeletek]

print("Teljesítmények:", teljesitmenyek)
print("Legnagyobb teljesítményű blokk indexe:", np.argmax(teljesitmenyek))

plt.figure(figsize=(8,4))
plt.bar(range(8), teljesitmenyek)
plt.title("1 másodperces hangblokkok teljesítménye")
plt.xlabel("Blokk index (0–7)")
plt.ylabel("Teljesítmény")
plt.tight_layout()
plt.show()
```

```python
# 7. Feladat – 1 perces adatok --> 10 perces blokkok diagramos ábrázolással együtt

import matplotlib.pyplot as plt

data = np.random.uniform(0.2, 1.5, size=600)  # 600 perc
blokkok = np.array_split(data, 60)            # 60 darab 10 perces blokk

blokkok_osszege = [b.sum() for b in blokkok]

plt.plot(blokkok_osszege)
plt.title("10 perces energiaösszeg")
plt.xlabel("Blokk index (0–59)")
plt.ylabel("Összeg")
plt.show()

print("Legnagyobb 10 perces fogyasztás indexe:", np.argmax(blokkok_osszege))
```

```python
# 7. feladat kicsit másfajta megjelenítése

plt.figure(figsize=(10,4))
plt.plot(blokkok_osszege, marker="o")
plt.title("10 perces időblokkok energiafogyasztása")
plt.xlabel("Blokk index (0–59)")
plt.ylabel("Összeg (kWh)")
plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()
```

```python
# 8. Feladat – 8×8 kép 4 patch-re bontása

img = np.random.randint(0, 256, size=(8, 8))

# Sorokra bontás
felso, also = np.array_split(img, 2, axis=0)

# Oszlopokra bontás
bal_fel   = np.array_split(felso, 2, axis=1)[0]
jobb_fel  = np.array_split(felso, 2, axis=1)[1]
bal_also  = np.array_split(also, 2, axis=1)[0]
jobb_also = np.array_split(also, 2, axis=1)[1]

patch_atlagok = [
    bal_fel.mean(),
    jobb_fel.mean(),
    bal_also.mean(),
    jobb_also.mean()
]

print("Patch-átlagok:", patch_atlagok)
print("Legvilágosabb patch indexe:", np.argmax(patch_atlagok))
```

```python
# 8. feladat kiegészítve diagramos ábrázolással

patches = [bal_fel, jobb_fel, bal_also, jobb_also]
titles = ["Bal-felső", "Jobb-felső", "Bal-alsó", "Jobb-alsó"]

plt.figure(figsize=(6,6))
for i, p in enumerate(patches):
    plt.subplot(2, 2, i+1)
    plt.imshow(p, cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.suptitle("Kép négy patch-re bontva")
plt.tight_layout()
plt.show()
```
